```python
#list of the element symbols from the periodic table up to 118
element_symbols = ["H", "He", "Li", "Be", "B", "C", "N", "O", "F", "Ne", 
"Na", "Mg", "Al", "Si", "P", "S", "Cl", "Ar", "K", "Ca", "Sc", "Ti", "V", 
"Cr", "Mn", "Fe", "Co", "Ni", "Cu", "Zn", "Ga", "Ge", "As", "Se", "Br", 
"Kr", "Rb", "Sr", "Y", "Zr", "Nb", "Mo", "Tc", "Ru", "Rh", "Pd", "Ag", 
"Cd", "In", "Sn", "Sb", "Te", "I", "Xe", "Cs", "Ba", "La", "Ce", "Pr", 
"Nd", "Pm", "Sm", "Eu", "Gd", "Tb", "Dy", "Ho", "Er", "Tm", "Yb", "Lu", 
"Hf", "Ta", "W", "Re", "Os", "Ir", "Pt", "Au", "Hg", "Tl", "Pb", "Bi", 
"Po", "At", "Rn", "Fr", "Ra", "Ac", "Th", "Pa", "U", "Np", "Pu", "Am", 
"Cm", "Bk", "Cf", "Es", "Fm", "Md", "No", "Lr", "Rf", "Db", "Sg", "Bh", 
"Hs", "Mt", "Ds", "Rg", "Cn", "Nh", "Fl", "Mc", "Lv", "Ts", "Og"]
abundances_url = "https://drive.google.com/file/d/1DgtjD0gAl_sCMmlr4pSyeu58HfAfZbuZ/view?usp=sharing"
download_file_g_drive_url(abundances_url, "/content/abundances.csv", False)
abundances_df = pd.read_csv("/content/abundances.csv")
def isotope_name(z_a):
  z,a = z_a[1:-1].split(",")
  symbol = element_symbols[int(z)-1]
  return a[1:] + symbol 
natural_z_a = [abundances_df.iloc[row,0] 
        for row in range(abundances_df.shape[0])
        if abundances_df.iloc[row,1]> 0]
natural_isotopes = [isotope_name(z_a) for z_a in natural_z_a]
```

```python
high_energy_isotopes = [col for col in 
df.columns if df.loc[0,col]>10000]
high_energy_isotopes
```

```python
natural_elements = [''.join([c for c in isotope if not c.isdigit()]) 
                          for isotope in natural_isotopes]
def is_natural_element(str_input):
  element = ''.join([c for c in str_input if not c.isdigit()]) 
  return element in natural_elements
def neutron_difference(str_input):
  if is_natural_element(str_input): 
    element = ''.join([c for c in str_input if not c.isdigit()]) 
    if element in natural_elements:
      a = int(''.join([c for c in str_input if c.isdigit()]))
      same_element_As = [''.join([c for c in isotope if c.isdigit()])
                        for isotope in natural_isotopes if element == 
                        ''.join([c for c in isotope if not c.isdigit()])]
      n_diffs = [int(a - int(A)) for A in same_element_As]
      return np.min(n_diffs)
  return "Artifical element"
```

```python
neutron_differences = [neutron_difference(isotope) 
                      for isotope in df.columns]
diff_dict = {}
for index in range(len(neutron_differences)):
  diff_dict[df.columns[index]] = neutron_differences[index]
if df.shape[0] == 1:
  df = df.append(diff_dict, ignore_index=True)
df
```

```python
keV_per_neutron_diff = {}
for isotope in df.columns:
  if not type(df.loc[1,isotope]) == str:
    if df.loc[1,isotope] == 0:
      print(isotope, " is natrually occuring\n")
    elif df.loc[1,isotope] < 0:
      print(isotope, " has fewer neutrons than naturally occuring")
    else:
      keV_per_neutron_diff[isotope] = df.loc[0,isotope] / df.loc[1, isotope]
keV_per_neutron_diff
```

```python
efficiency_df = pd.DataFrame(data = keV_per_neutron_diff, 
                                      index = ["Efficiency (keV per neutron)"])
efficiency_df = efficiency_df.transpose()
efficiency_df
```

```python
df.loc[0,"4H"]
```

```python
eff_isotopes = [x for x in efficiency_df.index]
efficiency_df["Total Energy (keV)"] = [df.loc[0, isotope] 
                                      for isotope in eff_isotopes]
efficiency_df["Parent Isotope"] = eff_isotopes
efficiency_df
```

```python
fig = px.scatter(efficiency_df, x="Total Energy (keV)", 
           y = "Efficiency (keV per neutron)", hover_data = ["Parent Isotope"],
           title = "Theoretical Maximum Efficiencies from Neutron Irradiation")
fig
```
