For every available isotope, plot additional energy released by the decay chain after a neutron is added to the nucleus.

```python
def energy(isotope):
    if isotope in df.columns:
        return df.loc[0,isotope]
    else:
        return 0
def add_neutron(isotope):
    element = ''.join([c for c in isotope if not c.isdigit()])
    a = int(''.join([c for c in isotope if c.isdigit()]))
    return str(a+1) + element
```

```python
post_absorption_energy = {}
for isotope in available_isotopes:
    post_absorption_energy[isotope] = energy(add_neutron(isotope)) - energy(isotope)
post_absorption_energy
```

```python
efficiency_df = pd.DataFrame(data = post_absorption_energy,
            index = ["Change in Energy Release with Neutron Absorption (keV)"])
efficiency_df = efficiency_df.transpose()
efficiency_df
```

Does this account for excited states in the nucleus?

```python
efficiency_df["Energy Release without Neutron Absorption (keV)"] = [
    energy(isotope) for isotope in efficiency_df.index
]
efficiency_df["Parent Isotope"] = [isotope for isotope in efficiency_df.index]
efficiency_df
```

```python
fig = px.scatter(efficiency_df, x="Energy Release without Neutron Absorption (keV)",
  y = "Change in Energy Release with Neutron Absorption (keV)",
  hover_data = ["Parent Isotope"],
  title = "Efficiency: Additional Energy Released by Decay Chain after Absorbing 1 Neutron")
fig
```

Now let's make the x-axis the half-life of the first decay in each chain, which will put a strict upperbound on the specific power of each isotope's decay chain.

```python
livechart.loc[livechart["Isotope"] == '84Nb', "half_life_sec"].unique()
```

```python
half_lives_dict = {}
isotopes_with_multiple_half_lives = []
isotopes_with_no_half_lives = []
for isotope in available_isotopes:
    half_lives = livechart.loc[livechart["Isotope"] == isotope, "half_life_sec"]
    unique_half_lives = half_lives.unique()
    if len(unique_half_lives) > 1:
        isotopes_with_multiple_half_lives.append(isotope)
    #if there are multiple half lives for an isotope
    #need to add more code to find an expectation value if this is the case
    if not all(unique_half_lives == ' '):
        half_lives_dict[isotope] = np.float64(unique_half_lives[0])
    else:
        isotopes_with_no_half_lives.append(isotope)
half_lives_dict
```

```python
[len(isotopes_with_no_half_lives),
 len(isotopes_with_multiple_half_lives), len(available_isotopes)]
```

So although we're missing 415 out of the 3367 half-lives, none of the isotopes in this version of livechart have multiple decay paths. Need to review why this is the case.

```python
known_isotopes = [isotope for isotope in efficiency_df.index if
    isotope not in isotopes_with_no_half_lives]
half_life_df = pd.DataFrame(data = half_lives_dict,
                            index = ["Half Life in Seconds"])
half_life_df = half_life_df.transpose()
half_life_df["Parent Isotope"] = [isotope for isotope in half_life_df.index]
half_life_df["Change in Energy Release after Neutron Absorption (keV)"] = [
    post_absorption_energy[isotope] for isotope in half_life_df.index]
half_life_df
```

```python
fig = px.scatter(half_life_df,
  x="Half Life in Seconds",
  y = "Change in Energy Release after Neutron Absorption (keV)",
  hover_data = ["Parent Isotope"],
  title = "Efficiency: Additional Energy Released by Decay Chain after Absorbing 1 Neutron",
                 log_x = True)
fig
```
