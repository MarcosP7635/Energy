```python
import pandas as pd
import numpy as np
import plotly.express as px
import glob
parent_dir = "C:\\Users\\engin\\Documents\\Github\\"
file_path = "C:\\Cross-Section-Data\\All_decay_modes_power_time_series_1001_steps.csv"
power_time_series = pd.read_csv(file_path)
power_time_series
```

```python
max_power_dict = {}
max_rows = [np.argmax(power_time_series[col]) for col in power_time_series.columns[:-1]]
max_power_dict["Power (watts)"] = [power_time_series.loc[max_rows[index], 
                                    power_time_series.columns[index]] 
                                    for index in range(len(power_time_series.columns[:-1]))]
max_power_dict["Time (seconds)"] = [float(time) for time in power_time_series.columns[:-1]]
max_power_dict["Parent Isotope"] = [power_time_series.loc[row, "Parent Isotope"] 
                                    for row in max_rows]
max_power_df = pd.DataFrame.from_dict(max_power_dict)
max_power_df
```

```python
px.scatter(max_power_df, y = "Power (watts)", x = "Time (seconds)", 
            color = "Parent Isotope", log_x = True, log_y = True)
```
