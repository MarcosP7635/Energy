# [Young, et al. P.P. 2013](https://doi.org/10.1016/j.phpro.2013.12.021)

This paper proposed a neutron source with a production rate of 1.41$\times$10$^8$ UCN/s/100W which is an energy efficiency of 1.41$\times$10$^6$ neutrons/J

```python
constants.N_A * units.J / (1.41 * 10**6)
```

This is off by a factor of a million

#[Thermofisher](https://www.thermofisher.com/search/browse/featured/us/en/80012997/Neutron%20Generators?resultPage=1)

MP320: Young et al.'s 2013 paper

```python
constants.N_A * units.J / (10**8 / 50)
```

API 120: Also same as Young et al.'s 2013 paper

```python
I = 50 * units.microamp
V = 90 * units.kV
E = (V * I * units.second).to(units.J)
constants.N_A / (10**7 / E)
```
