```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib notebook #incompatible with mpmath
import mpmath as m #incompatible with matplotlib notebook
m.dps= 10**3
def half_life_to_energy(half_life, time, initial_mass):
    #we assume half life and time are in seconds
    #mass is in kg
    c = 3 * m.power(10, 8) #meters per second
    mass_decayed = mass_lost(half_life, time, initial_mass)
    energy = mass_decayed * m.power(c, 2)
    return (energy, mass_decayed) #joules, kg
    
'''
Because of this function, do not name anything else mass_lost !!
'''
def mass_lost(half_life, time, initial_mass):
    mass_lost = initial_mass * (1- m.power(.5, (time / half_life)))
    return mass_lost

def convert_s_to_y(time):
    years = time / 31536000
    return years

half_life_U_238 = 1.41* m.power(10, 17) #seconds 
half_life_CA_48 = 6.4 * m.power(10, 26.5) #seconds
#around 10^9.5 years :)
half_life = m.power(10, 11)
mass = m.power(10, -6) #kilograms
time = 1 #second
power, mass_decayed = half_life_to_energy(half_life,                              
            time, mass)
print(mass, "kg generates", power, "watts and loses", mass,
      "kg of mass over the first\nsecond :).\nIt will lose" +
      " half its mass in", convert_s_to_y(half_life), "years")
```

```python
'''
Here we will plot half-life versus power generated
'''
precise_half_lives, power_array = [], []
for exponent in range(1, 15):
    half_life = m.power(10, exponent)
    precise_half_lives.append(half_life)
    
    power, mass_decayed = half_life_to_energy(half_life,                              
            time, mass)
    power_array.append(power)

plt.figure(dpi = 140)
plt.scatter(precise_half_lives, power_array)
plt.xscale("log"), plt.yscale("log")
plt.title("Power Produced by " + str(mass) + " kg of Different Isotopes")
plt.ylabel("Power in Watts")
plt.xlabel("Half Life in Seconds")
plt.grid(which='major', color='black')
plt.grid(which='minor', color='grey', alpha=0.4)
plt.minorticks_on()
```
